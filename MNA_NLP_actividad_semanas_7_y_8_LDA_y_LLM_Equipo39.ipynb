Se eligió un LLM en lugar de corrección por regex porque los errores de transcripción

¿Qué es un LLM?

Un **LLM (Large Language Model)** o Modelo de Lenguaje Grande es un tipo de 
inteligencia artificial entrenado con enormes cantidades de texto para entender 
y generar lenguaje humano.

¿Cómo funciona?
Durante su entrenamiento, el modelo aprende patrones del lenguaje — gramática, 
contexto, significado — a partir de millones de textos. Esto le permite:
- Entender el contexto de una oración
- Predecir qué palabra tiene más sentido en un lugar dado
- Corregir errores manteniendo el significado original

¿Por qué usarlo para corregir transcripciones?
A diferencia de un corrector ortográfico tradicional (que solo revisa palabras 
de forma aislada), un LLM entiende el contexto completo. Por ejemplo:
- Sabe que "logo" en una fábula de animales debe ser "lobo"
- Sabe que "sora" en ese contexto debe ser "zorra"
- No cambia palabras correctas que parecen raras fuera de contexto

# **Ejercicio 4:**

De cada fábula deberán extraer las palabras clave mediante el algoritmo LDA
(Asignación Latente de Dirichlet). Consideren que cada fábula tiene solo un tópico y
que el total de palabras por tópico serían en principio 20 (pero pueden ajustar este
valor si consideran que obtienen un mejor resultado). Deberán desplegar las
palabras clave obtenidas en cada fábula.

```python
import json
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from gensim import corpora
from gensim.models import LdaModel

# Descarga recursos necesarios de NLTK:
# - 'stopwords': listas de palabras vacías (artículos, preposiciones, etc.)
# - 'punkt' / 'punkt_tab': modelos para dividir el texto en palabras (tokenizar)
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')

# Carga el conjunto de stopwords en español (ej: "el", "la", "de", "que"...)
stop_words = set(stopwords.words('spanish'))

def preprocesar(texto):
    """
    Limpia y tokeniza un texto para análisis de tópicos:
    - Convierte todo a minúsculas y separa en palabras (tokens).
    - Conserva solo tokens que sean palabras (isalpha, sin números/símbolos),
      que NO sean stopwords, y que tengan más de 3 letras
      (para descartar palabras muy cortas poco informativas).
    """
    tokens = word_tokenize(texto.lower())
    tokens = [t for t in tokens if t.isalpha() and t not in stop_words and len(t) > 3]
    return tokens

# --- Preprocesa cada fábula y guarda los tokens resultantes ---
documentos = {}      # diccionario: clave -> lista de tokens de esa fábula
corpus_tokens = []   # lista de listas de tokens (una lista por fábula), insumo para gensim

for clave, texto in transcripciones_corregidas.items():
    tokens = preprocesar(texto)
    documentos[clave] = tokens
    corpus_tokens.append(tokens)

# --- Construye el vocabulario y representación numérica del texto ---
# El diccionario asigna un ID único a cada palabra del vocabulario completo
diccionario = corpora.Dictionary(corpus_tokens)

# Convierte cada fábula a formato "bag of words" (BoW): lista de pares
# (id_palabra, frecuencia) según el diccionario anterior
corpus_bow = [diccionario.doc2bow(tokens) for tokens in corpus_tokens]

# Lista de claves (nombres de las fábulas) en el mismo orden que corpus_bow
claves = list(transcripciones_corregidas.keys())

# Diccionario donde se guardarán las palabras clave de cada fábula,
# para reutilizarlas en pasos posteriores
lda_keywords = {}

print("=" * 60)
print("PALABRAS CLAVE POR FÁBULA (LDA)")
print("=" * 60)

# --- Entrena un modelo LDA independiente para cada fábula ---
# (num_topics=1 porque se busca UN solo "tema" que resuma cada fábula
# por separado, no temas compartidos entre todas las fábulas)
for i, (clave, bow) in enumerate(zip(claves, corpus_bow)):
    lda = LdaModel(
        corpus=[bow],          # solo el BoW de esta fábula
        id2word=diccionario,   # vocabulario compartido
        num_topics=1,          # un único tópico por fábula
        passes=50,             # número de iteraciones de entrenamiento (más = más estable)
        random_state=42        # semilla fija para resultados reproducibles
    )
    
    # Obtiene las 20 palabras más representativas del tópico, con su probabilidad
    palabras = lda.show_topic(0, topn=20)
    
    # Se queda solo con las palabras, descartando las probabilidades
    keywords = [palabra for palabra, prob in palabras]
    
    # Guarda las palabras clave de esta fábula para uso posterior
    lda_keywords[clave] = keywords
    
    # Muestra el resultado de esta fábula
    print(f"\n {clave}:")
    print(", ".join(keywords))
    print("-" * 60)
```

Por ello, esta actividad no solo tiene valor académico, sino que refleja  herramientas centrales en la transformación digital actual:

- **Accesibilidad**: la transcripción automática de audio facilita el acceso a contenido para personas con discapacidad auditiva, y permite digitalizar y preservar archivos sonoros y patrimonio cultural (como las fábulas de  Esopo en este caso).

- **Reuniones de trabajo y juntas corporativas**: el mismo flujo audio→texto permite generar actas y minutas automáticamente. La transcripción automatizada de reuniones ayuda a que tanto los asistentes como quienes no pudieron asistir aprovechen al máximo cada reunión,  evitando que esa información se pierda.

- **Educación y conferencias**: el software de transcripción ofrece resultados de alta fidelidad para reuniones, conferencias y entrevistas, convirtiéndose en un recurso indispensable para educadores,  permitiendo crear apuntes, repositorios de clases y materiales de estudio searchables.

- **Ámbito jurídico**: el sector legal está aumentando el uso de software de dictado y transcripción, útil para redactar documentos legales y tomar notas de reuniones,  agilizando procesos que antes requerían transcripción manual.

- **Salud**: plataformas de transcripción de reuniones y dictado cumplen normativas al manejar datos médicos confidenciales,  lo que demuestra cómo estas tecnologías ya se aplican a la digitalización segura de información clínica hablada.

- **Investigación cualitativa y análisis de entrevistas**: herramientas similares permiten comparar temas en decenas de entrevistas y transformar conversaciones con clientes o usuarios en resúmenes y materiales visuales,  un proceso análogo al de extraer palabras clave por fábula mediante LDA en esta actividad.

Estos casos muestran que el pipeline desarrollado  no es exclusivo del análisis literario, sino una base aplicable a prácticamente cualquier contexto donde se genere información hablada que deba convertirse en texto útil, preciso y analizable.
